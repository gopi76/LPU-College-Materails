```python
import cv2
import numpy as np
from google.colab.patches import cv2_show
```

```python
img = cv2.read('pancard.png')   #loading  the img

#convert input img into grayscale img

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray = np.float32(gray)
```

```python
corners = cv2.cornerHarris(gray, 2, 3, 0.05)
```

```python
corners = cv2.dilate(corners, None)
```

```python
img[corners > 0.01 * corners.max()] = [0, 0, 255]
```

```python
cv2_imshow(img)
```
