```python
import numpy as np

data = np.array([100, 120, 100, 150, 160])
mask = np.array([1, 0, -1])
```

```python
conv = np.convolve(data, mask, 'full')
corr = np.correlate(data, mask, 'full')

print("Convultion : ", conv)
print("Correlation : ", corr)
```

# Program of Average Filter

```python
import cv2
from matplotlib import pyplot as plt

#load an image
img = cv2.imread('pancard.png', cv2.IMREAD_GRAYSCALE)

#apply the average filter
kernel_size = 5
filtered_img = cv2.blur(img, (kernel_size, kernel_size))

plt.figure(figsize = (10, 6))

plt.imshow(filtered_img)
plt.title("filtered img")
```

# Program for Median Blur

```python

blurred_img = cv2.medianBlur(img, kernel_size)


plt.figure(figsize = (10, 6))

plt.imshow(blurred_img, cmap="gray")
plt.title("median blur img")
```

```python
#apply max filter

kernel = np.ones((kernel_size, kernel_size), np.uint8)

max_filter = cv2.dilate(img, kernel)

plt.figure(figsize = (10, 6))

plt.imshow(max_filter, cmap = "gray")
plt.title(" max- filtered img")
```

```python
#apply the min_filtered img


min_filter = cv2.erode(img, kernel)

plt.imshow(min_filter, cmap = "gray")
```

```python
#apply gussain blur

kernel_size = (5, 5)
sigma_x     = 0

gussian_blur = cv2.GaussianBlur(img, kernel_size, sigmaX = sigma_x)

plt.figure(figsize= (10,6))

plt.imshow(gussian_blur)
plt.title("Gaussian Blur image")
```

```python
#apply bilateral img

diameter = 9
sigma_color, sigma_space = 75, 75

bilateral_filter = cv2.bilateralFilter(img, diameter, sigma_color, sigma_space)

plt.figure(figsize=(10, 6))
plt.imshow(bilateral_filter)
```

```python

laplacian = cv2.Laplacian(img)  #used for sharpening the img

plt.imshow(laplacian)
```

```python

#program for translation


image_path ="pancard.png"

original = cv2.imread(image_path)

original1 = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

plt.imshow(original1)
plt.show()
```

```python
tx, ty = 50, 30

translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

translated_img = cv2.warpAffine(original1, translation_matrix, (original1.shape[1], original1.shape[0]))

plt.imshow(translated_img)
plt.axis('off')
plt.show()
```

```python
#program for scaling


scale_x, scale_y = 1, 0.6

new_width = int(original1.shape[1] * scale_x)
new_height = int(original1.shape[0] *scale_y)

scaled_img = cv2.resize(original1, (new_width, new_height))

plt.imshow(scaled_img)
plt.show()
```
